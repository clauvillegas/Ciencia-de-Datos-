```python
#Definicion de funciones

def validanumero(tipo,mensaje):
    while True:
        try:
            if tipo == "int":
              numero = int(input(mensaje))
            elif tipo == "float":
              numero = float(input(mensaje))
            break  # Salir del bucle si el input es válido
        except ValueError:
            print("Error: Debes ingresar un número válido.")
    return numero

#-------------------------------------------------------------------------------------------
#-------------------------------------------------------------------------------------------


def filtrar_libros_por_precio(libros, rango_ini, rango_fin):
    libros_filtrados = [libro for libro in libros if rango_ini <= libro["precio"] <= rango_fin]
    return libros_filtrados
#-------------------------------------------------------------------------------------------
#-------------------------------------------------------------------------------------------

def filtrar_libros_por_disponibilidad(libros, cant):
    libros_filtrados = [libro for libro in libros if libro["cantidad_en_stock"] > cant]
    return libros_filtrados
#-------------------------------------------------------------------------------------------
#-------------------------------------------------------------------------------------------

# 2. y 3. Muestro todos los disonibles cuando ini == 0 y fin == 0
def mostrar_libros(ini, fin, libros, que_libros):
    print("\n")
    #filtra todos los libros (que_libros=T) o solo con stock>0  (que_libros=D)
    if que_libros.lower() == "d":
        libros_filtrados = filtrar_libros_por_disponibilidad(libros, 0)
        tipo = " disponibles "
    else:
        libros_filtrados = libros
        tipo = " "
    #filtra por rango de precio (si ini y fin son 0, no filtra )
    if (ini==0 and fin==0):
        print("Libros"+tipo+":")
    else:
        print(f"Libros"+tipo+"con Precio entre $" + str(ini) + "y $" + str(fin) + ":")
        libros_filtrados = filtrar_libros_por_precio(libros_filtrados,ini,fin)
    print("-" * 90)
    #muestra listado
    if not libros_filtrados:
      print("No hay libros disponibles.")
    else:
      print(f"{'Título':<40}|{'Autor':30}|{'Precio':>8}|{'Stock':>6}")
      print("-" * 90)
      for libro in libros_filtrados:
        print(f"{libro['titulo']:<40}|{libro['autor']:30}|{libro['precio']:>8}|{libro['cantidad_en_stock']:>6}" )

    print("-" * 90)


#-------------------------------------------------------------------------------------------
#-------------------------------------------------------------------------------------------

# funcion comprar
def comprar_libros(libros,libro_dsctos):
  encontrado = False
  #libros_lista = []
  dscto = 0

  print("\n")
  print("\033[4mComprar libros:\033[0m\n")

  titulo = input("Ingrese el título del libro a comprar: ")
  cantidad = validanumero("int","Ingrese la cantidad a comprar: ")

  print("\n")

  for libro in libros:
    if libro["titulo"] == titulo:
      autor= libro["autor"]
      precio = libro["precio"]
      stock = libro["cantidad_en_stock"]
      encontrado = True
      break

  if not encontrado:
    print(f"No se encontró un libro con el título '{titulo}'.")
  else:
      #revisa descuento autor
      dscto = 0

      for dscto_autor in libro_dsctos:
        if dscto_autor["autor"] == autor:
          dscto = dscto_autor["descuento"]
          break

      #revisa stock disponible
      if stock >= cantidad:
        libro["cantidad_en_stock"] -= cantidad
        valor_compra = libro["precio"] * cantidad * ((100 - dscto)/100)

        if dscto > 0:
          print(f"Se han comprado {cantidad} unidades de '{titulo}' por un valor de $ " + str(int(valor_compra)) + ". (valor - " + str(dscto) + "% descuento)")
        else:
          print(f"Se han comprado {cantidad} unidades de '{titulo}' por un valor de $ " + str(int(valor_compra)) + ". ")

      else:
        print(f"Stock insuficiente.")
  return {"titulo": titulo, "cantidad": cantidad, "descuento": dscto, "valor_compra": valor_compra}

#-------------------------------------------------------------------------------------------
#-------------------------------------------------------------------------------------------
# muestra libros comprados
def mostrar_libros_comprados(libros_factura):
  print("\n")
  print("Libros Comprados")
  i=1
  valor_total = 0

  print("-" * 90)
  for libro in libros_factura:
    print(f"{libro['cantidad']:<5} x {libro['titulo']:<40} - Dscto:{libro['descuento']:<5}% => ${libro['valor_compra']:>20}" )
    valor_total += libro['valor_compra']
  print("-" * 90)
  print("\n")

  print(f"{'Total Compra  =>'.rjust(75)} ${valor_total:>10}" )


#-------------------------------------------------------------------------------------------
#-------------------------------------------------------------------------------------------

# muestra libros por rango de precio
def libros_rango(libros):
  while True:
    print("\n")
    print("\033[4mFiltrar libros por rango de precio:\033[0m\n")
    rango_ini = validanumero("int","Ingrese precio mas bajo: ")
    rango_fin = validanumero("int","Ingrese precio mas alto: ")
    if rango_ini > rango_fin:
      print("El precio mas bajo no puede ser mayor al mas alto.")
    else:
      break

  print("\nQuieres una lista de todos los libros o solo los disponibles? ")
  while True:
    que_libros = input("T: todos/ D:disponibles): ")
    if que_libros.lower() == "d" or que_libros.lower() == "t":
      break
    else:
      print("Por favor, ingrese 'T' o 'D'.")

  mostrar_libros(rango_ini, rango_fin, libros, que_libros)

#-------------------------------------------------------------------------------------------
#-------------------------------------------------------------------------------------------

# graba descuento
def grabar_descuento(d_autor):
    while True:
      print("\n")
      autor = input("Ingrese el Autor: ")
      while True:
        valor_descuento = validanumero("float","Descuento: ")
        if 1<=valor_descuento<= 100:
          dato_descuento= {"autor": autor, "descuento": valor_descuento}
          d_autor.append(dato_descuento)
          break
        else:
          print("El descuento debe estar entre 1 y 100.")
      salir = input("Desea continuar? (s/n): ")
      if salir.lower() == "n":
          break
    return d_autor
```

```python

#1. Definir variables básicas y tipos de datos
libros = []
datos_libro = {}
descuento_autor=[]
libros_comprados = []
```

```python

#o Crea una lista que contenga al menos cinco libros, donde cada libro sea un diccionario con los atributos título (cadena de caracteres), autor (cadena de caracteres), precio (decimal) y cantidad en stock (entero)

#Opcion de ingresar libros si es que no estan ya creados
stock = input("Desea ingresar stock de libros? (s/n): ")

if stock.lower() == "s":
  while True:
      print("\n")
      print("\033[4mIngrese los datos del libro:\033[0m\n")
      titulo = input("Titulo: ")
      autor = input("Autor: ")
      precio = validanumero("float","Precio: ")
      cant = validanumero("int","Stock: ")
      salir = input("Desea continuar? (s/n): ")
      datos_libro = {"titulo": titulo, "autor": autor, "precio": precio, "cantidad_en_stock": cant}
      libros.append(datos_libro)
      if salir.lower() == "n":
          break
  print("\n")

mostrar_libros(0,0, libros, "D")
```

```python
#graba descuentos por autor
descuentos = input("Desea ingresar descuento por Autor (s/n): ")
if descuentos.lower() == "s":
  descuento_autor = grabar_descuento(descuento_autor)
  print("\n")
```

```python
while True:
  print("----------- Sistema de Compras -----------\n")
  print("1. Mostrar libros disponibles")
  print("2. Filtrar libros por rango de precio")
  print("3. Comprar libros")
  print("4. Finalizar Compra y Mostrar Factura")

  # Valida opcion ingresada
  while True:
    opcion = validanumero("float", "Ingrese una opción: ")
    if 1<=opcion<=4:
      break
    else:
      print("Opción Inválida\n")

  #Ejecuta accion segun Opcion
  if opcion == 1:
     mostrar_libros(0,0, libros, "D")
  elif opcion == 2:
    libros_rango(libros)
  elif opcion == 3:
    while True:
      libros_comprados.append(comprar_libros(libros, descuento_autor))
      salir = input("Desea continuar? (s/n): ")
      if salir.lower() == "n":
          break
  elif opcion == 4:
    mostrar_libros_comprados(libros_comprados)
    break

```
